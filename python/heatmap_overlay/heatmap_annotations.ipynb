```python
import numpy as np
import imageio
import matplotlib.pyplot as plt

from skimage.color import rgb2gray
from skimage.color import label2rgb
```

```python
img = imageio.imread('imgs/image-4138.png')
plt.imshow(img, cmap="gray")
plt.title("Original image")
plt.show()
```

```python
heatmap = imageio.imread('preds/prediction-4134.png')

plt.imshow(heatmap)
plt.title("Heatmap")
plt.show()
```

```python
hmap = heatmap[:,:,0]  # we have heatmap given with same value in every color channel

plt.imshow(hmap, cmap="jet", vmin=0, vmax=255)
plt.title("Heatmap")
plt.show()
```

```python
plt.imshow(img, cmap='gray', interpolation='none') # I would add interpolation='none'
plt.imshow(hmap, cmap='jet', vmin=0, vmax=255, alpha=0.5, interpolation='none') # interpolation='none'
plt.title("Overlay")
plt.show()
```

```python
fig = plt.figure(frameon=False)
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(img, cmap='gray', interpolation='none')
ax.imshow(hmap, cmap='jet' , vmin=0, vmax=255, alpha=0.5, interpolation='none')
plt.savefig("overlay_4134.png")
plt.show()
```

```python
tups = [
    ("imgs/image-4134.png", "preds/prediction-4134.png"),
    ("imgs/image-4135.png", "preds/prediction-4135.png"),
    ("imgs/image-4136.png", "preds/prediction-4136.png"),
    ("imgs/image-4137.png", "preds/prediction-4137.png"),
    ("imgs/image-4138.png", "preds/prediction-4138.png"),
]
```

```python
for tup in tups:
    num = tup[0][-8:-4]
    img = imageio.imread(tup[0])
    heatmap = imageio.imread(tup[1])
    hmap = heatmap[:,:,0]
    
    fig = plt.figure(frameon=False)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    ax.imshow(img, cmap='gray', interpolation='none')
    ax.imshow(hmap, cmap='jet', alpha=0.5, interpolation='none')  # vmin=0, vmax=255 to normalize...
    plt.savefig("heatmap_{}.png".format(num))
    plt.show()
```
